### Mechine learning menggunakan Decesion Tree Classifier (Kelompok 1) 

```python
X_DTC = df[['Radius (mean)', 'Texture (mean)',
        'Perimeter (mean)', 'Area (mean)', 
        'Smoothness (mean)','Compactness (mean)',
        'Concavity (mean)', 'Concave points (mean)',
        'Symmetry (mean)', 'Fractal dimension (mean)',
        ]]
y_DTC = df['Diagnosis']

X_DTC_train, X_DTC_test, y_DTC_train, y_DTC_test = train_test_split(X_DTC, y_DTC, test_size=0.2, random_state=42)
```

```python
# Membuat model Decision Tree Classifier
model_DTC = DecisionTreeClassifier()

# Melatih model dengan data train
model_DTC.fit(X_DTC_train, y_DTC_train)

# Memprediksi target dari data test
y_DTC_prediksi = model_DTC.predict(X_DTC_test)
```

```python
# Menghitung akurasi dari model
accuracy_DTC = accuracy_score(y_DTC_test, y_DTC_prediksi)
print("Akurasi:", accuracy_DTC)
```

## Membuat Machine Learning 

```python
df= pd.read_csv(r"D:\DATA PELATIHAN\cancer.csv")

df
```

### Machine Learning menggunakan model Random Forres (kelompok 1)

```python
X_RF = df[['Radius (mean)', 'Texture (mean)','Perimeter (mean)', 'Area (mean)','Smoothness (mean)','Compactness (mean)',
           'Concavity (mean)', 'Concave points (mean)','Symmetry (mean)', 'Fractal dimension (mean)']]
y_RF = df['Diagnosis']

X_RF_train, X_RF_test, y_RF_train, y_RF_test = train_test_split(X_RF, y_RF, test_size=0.2, random_state=42)
```

```python
# Membuat model randomForesClassifier
model_RF = RandomForestClassifier(n_estimators=300, random_state=42)

#melatih model menggunakan data train
model_RF.fit(X_RF_train, y_RF_train)

# membuat prediksi menggunakan data test
y_RF_predict = model_RF.predict(X_RF_test)

# menghiung akurasi model
accuracy_RF = accuracy_score(y_RF_test, y_RF_predict)
print('Accuracy:', accuracy_RF)
```
